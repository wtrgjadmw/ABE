```python
def ep_mul_naf_(p:ep_t,k:int)->ep_t: #ep_mul_naf_imp
    EP_WIDTH=4
    tab=[None]*EP_WIDTH
    tab[0]=ep_dbl(p)
    tab[1]=ep_add(tab[0],p)      # = 3p
    tab[2]=ep_add(tab[1],tab[0]) # = 5p
    tab[3]=ep_add(tab[2],tab[0]) # = 7p
    tab[0]=p                     # = p
    for i in range(1,EP_WIDTH):
        tab[i]=ep_norm(tab[i])

    _k = k % param.r
    naf,naf_len=bn_rec_naf(383,_k,EP_WIDTH)
    
    r=ep_t()
    for i in range(naf_len,0,-1):
        r=ep_dbl(r)
        u=naf[i-1]
        if(u>0):
            r=ep_add(r,tab[int(u/2)])
        elif(u<0):
            r=ep_sub(r,tab[int(-u/2)])

    r=ep_norm(r)
    if(k<0):
        r=ep_neg(r)
    return r
```

```python
p=ep_rand()
r=ep_mul(p,25)
r2=ep_mul_naf_(p,25)
ep_print(r)
ep_print(r2)
```

```python
def ep2_mul_naf_(p:ep2_t,k:int)->ep2_t: #ep2_mul_naf_imp
    EP_WIDTH=4
    tab=[None]*EP_WIDTH
    tab[0]=ep2_dbl(p)
    tab[1]=ep2_add(tab[0],p)      # = 3p
    tab[2]=ep2_add(tab[1],tab[0]) # = 5p
    tab[3]=ep2_add(tab[2],tab[0]) # = 7p
    tab[0]=ep2_cpy(p)             # = p
    
    for i in range(1,EP_WIDTH):
        tab[i]=ep2_norm(tab[i])

    _k = k % param.r
    naf,naf_len=bn_rec_naf(383,_k,EP_WIDTH)

    r=ep2_t()
    for i in range(naf_len,0,-1):
        r=ep2_dbl(r)
        u=naf[i-1]
        if(u>0):
            r=ep2_add(r,tab[int(u/2)])
        elif(u<0):
            r=ep2_sub(r,tab[int(-u/2)])

    r=ep2_norm(r)
    if(k<0):
        r=ep2_neg(r)
    return r
```

```python
p=ep2_rand()
r=ep2_mul(p,25)
r2=ep2_mul_naf_(p,25)
print("r")
ep2_print(r)
print("r2")
ep2_print(r2)
```
